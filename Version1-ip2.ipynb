```python
#Deriving analogous colors from starting rgb values

sr, sg, sb = 50, 50, 170 #original rgb

def analogous(r,g,b):
    newr1 = r
    newg1 = g + 31 #get new values
    if newg1 > 255:
        newg1 = g - 31 #make sure new values aren't higher than 255
    newb1 = b

    newr2 = r
    newg2 = g + 16
    if newg2 > 255:
        newg2 = g - 16
    newb2 = b

    return [newr1, newg1, newb1, newr2, newg2, newb2]

analogous = analogous(sr,sg,sb) #get analogous values
print(analogous)

```

```python
#Deriving tetrad colors from starting rgb values

sr, sg, sb = 50, 50, 170 #original rgb

def tetrad(r,g,b):
    newr1 = r + 71
    if newr1 > 255:
        newr1 = g - 71 #make sure new values aren't higher than 255
    newb1 = b
    newg1 = g - 46
    if newg1 < 0:
        newg1 = g + 46 #make sure new values aren't lower than 0
    newb1 = b
    newb1 = b

    newr2 = r + 78
    if newr2 > 255:
        newr2 = g - 78
    newg2 = g - 14
    if newg2 < 0:
        newg2 = g + 14
    newb2 = b - 78
    if newb2 < 0:
        newb2 = g + 78

    return [newr1, newg1, newb1, newr2, newg2, newb2]


tetrad = tetrad(sr, sg, sb)
print(tetrad)
```

```python
def force_bullet(paragraph, char="â€¢"):
    """
    Ensure the paragraph in a textbox shows a bullet.
    """
    pPr = paragraph._p.get_or_add_pPr()
    # Remove explicit 'no bullet'
    for el in pPr.findall(qn('a:buNone')):
        pPr.remove(el)
    # Add bullet char if not present
    if not pPr.findall(qn('a:buChar')):
        buChar = OxmlElement('a:buChar')
        buChar.set('char', char)
        pPr.append(buChar)
```
